**Traditonal ML Algorithm**


* Random forest
* SVM
* Naive Bayes


```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn import svm
from sklearn.naive_bayes import GaussianNB

import tensorflow as tf
from tensorflow.keras import layers
#import tensorflow_hub as hub
import tensorflow_datasets as tfds

import numpy as np
import pandas as pd
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
```

```python
 #Load the dataset
(ds_train, ds_test), ds_info = tfds.load(
    'cats_vs_dogs',
    split=['train[:80%]', 'train[80%:]'],
    shuffle_files=True,
    as_supervised=True,
    with_info=True,
)

# Define the batch size and image size
BATCH_SIZE = 32
IMG_SIZE = 224

# Preprocess the data
def preprocess(image, label):
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = tf.cast(image, tf.float32) / 255.0
    label = tf.cast(label, tf.int32)
    return image, label

ds_train = ds_train.map(preprocess).batch(BATCH_SIZE)
ds_test = ds_test.map(preprocess).batch(BATCH_SIZE)
```

```python
train_cl=[]
test_cl=[]
for i,j in ds_train:
    train_cl.append(j)
train_cl = tf.concat(train_cl, axis=0)
for i,j in ds_test:
    test_cl.append(j)
test_cl=tf.concat(test_cl, axis=0)
```

```python
# Extract features using a pre-trained CNN
base_model = tf.keras.applications.ResNet50(include_top=False, weights='imagenet', pooling='avg')
train_features = base_model.predict(ds_train, verbose=1)
test_features = base_model.predict(ds_test, verbose=1)
```

**RANDOM FOREST**

```python
# Train a random forest classifier on the features
clf = RandomForestClassifier(n_estimators=100)
clf.fit(train_features, train_cl)

# Evaluate the performance on the test set
predictions = clf.predict(test_features)
accuracy = accuracy_score(test_cl, predictions)
print("Accuracy:", accuracy)
```

**SVM**

```python
# Train a random forest classifier on the features
clf = svm.SVC()
clf.fit(train_features, train_cl)

# Evaluate the performance on the test set
predictions = clf.predict(test_features)
accuracy = accuracy_score(test_cl, predictions)
print("Accuracy:", accuracy)
```

**Naive Bayes**

```python

# Train a Naive Bayes classifier on the features
clf = GaussianNB()
clf.fit(train_features, train_cl)

# Evaluate the performance on the test set
predictions = clf.predict(test_features)
accuracy = accuracy_score(test_cl, predictions)
print("Accuracy",accuracy)
```
